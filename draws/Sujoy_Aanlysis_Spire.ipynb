```pyspark3
spire = spark.read.load('s3://sita-coe-ds-dev-v1/mart/spire/airsafe')
```

```pyspark3
spire.registerTempTable('spire')
```

```pyspark3
spire.show(5)
```

```pyspark3
spark.sql("select timestamp, latitude, longitude, scheduled_departure_time_utc, scheduled_arrival_time_utc, tail_number, ingestion_time, record_ingest_time, record_ingest_date, record_ingest_hour from spire where record_ingest_date = '2019-08-16' and callsign = 'AAL2348' order by timestamp").show(100)
```

```pyspark3
spark.sql("select * from spire where record_ingest_date = '2019-08-18' and flight_number = 'UA746' order by timestamp desc").show(500)
```

```pyspark3
spark.sql("select count(distinct callsign) from spire where record_ingest_date = '2019-08-18' and origin_airport_icao = 'KJFK'").show(1000)
spark.sql("select callsign, count(*) from spire where record_ingest_date = '2019-08-18' and origin_airport_icao = 'KJFK' group by callsign").show(1000)
```

```pyspark3
spark.sql("select count(distinct callsign) from spire where record_ingest_date = '2019-08-18' and origin_airport_icao = 'KSFO'").show(1000)
spark.sql("select callsign, count(*) from spire where record_ingest_date = '2019-08-18' and origin_airport_icao = 'KSFO' group by callsign").show(1000)
```
